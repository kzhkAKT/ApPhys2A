дёҖйӣ»иҚ·гҒӮгҒҹгӮҠгҒ«еғҚгҒҸд»•дәӢгҒҜ
- йӣ»е ҙ $E$ дёӯгҒ«еғҚгҒҸйӣ»иҚ· $e$ гҒ®еҠӣ $f$: $f=e\,E$
- йӣ»жөҒгҒ®(е№іеқҮ)йҖҹеәҰгӮ’$v$, иЁҲжё¬жҷӮй–“гӮ’ $t$ гҒЁгҒҷгӮӢгҒЁз§»еӢ•и·қйӣў $\ell$: $\ell = v\, t$

\begin{align*}
w & = f\,\ell = \left( e\,E \right) (v\,t)
\end{align*}

- е°Һз·ҡгҒ®й•·гҒ•гӮ’$L$, гҒқгҒ“гҒ« йӣ»дҪҚе·® $V$ гҒҢгҒӮгӮҢгҒ°
  - еҚҳдҪҚйӣ»иҚ·гҒ«еҜҫгҒ—гҒҰгҖҒеҚҳдҪҚй•·гҒӮгҒҹгӮҠгҒ®йӣ»дҪҚе·®гҒҢ $E = \dfrac{V}{L}$
- е°Һз·ҡгҒ®й•·гҒ•$L$гҒ«гҒҜгҖҒ$N=\,n\,S\,L$ еҖӢгҒ®йӣ»иҚ·гҒҢгҒӮгӮӢ

е°Һз·ҡеҶ…гҒ§йӣ»е ҙгҒӢгӮүгҒӘгҒ•гӮҢгӮӢд»•дәӢгҒҜ

\begin{align*}
W &= N\,w = e n\,S\,L\left(\dfrac{V}{L} \right) (v\,t)\\
&= e\,n\,S\,v\cdot V t = I V t\\
&= I (I\,R) t \\[2mm]
W &= I^2 R t
\end{align*}

#### е°Һз·ҡеҶ…гҒ§ж¶ҲиІ»гҒ•гӮҢгӮӢд»•дәӢзҺҮ $P=\dfrac{W}{t}$

\begin{align*}
P &= \dfrac{W}{t} \\[1mm]
&= \dfrac{I V t}{t} = I\,V\\
&= IV = I\,(I\,R)= I^2\,R\\
&= IV = \frac{V}{R}\,V = \dfrac{V^2}{R}
\end{align*}
