一電荷あたりに働く仕事は
- 電場 $E$ 中に働く電荷 $e$ の力 $f$: $f=e\,E$
- 電流の(平均)速度を$v$, 計測時間を $t$ とすると移動距離 $\ell$: $\ell = v\, t$

\begin{align*}
w & = f\,\ell = \left( e\,E \right) (v\,t)
\end{align*}

- 導線の長さを$L$, そこに 電位差 $V$ があれば
  - 単位電荷に対して、単位長あたりの電位差(印加**電圧**)が $E = \dfrac{V}{L}$
- 導線の長さ$L$には、$N=\,n\,S\,L$ 個の電荷がある

導線内で電場からなされる仕事は

\begin{align*}
W &= N\,w = e n\,S\,L\left(\dfrac{V}{L} \right) (v\,t)\\
&= e\,n\,S\,v\cdot V t = I V t\\
&= I (I\,R) t \\[2mm]
W &= I^2 R t
\end{align*}

#### 導線内で消費される仕事率 $P=\dfrac{W}{t}$

\begin{align*}
P &= \dfrac{W}{t} \\[1mm]
&= \dfrac{I V t}{t} = I\,V\\
&= IV = I\,(I\,R)= I^2\,R\\
&= IV = \frac{V}{R}\,V = \dfrac{V^2}{R}
\end{align*}

---

本来であれば、電圧は次のように定義されるが
$$
E_{\rm true} =  -  \dfrac{V-0}{L-0} =  -  \dfrac{\varDelta V}{\varDelta L}
$$
この教科書では、その大きさだけに注目しているため
$$
E = \left| E_{\rm true} \right|= V/L
$$
で良い。

また、仕事率についても、丁寧に書くのであれば
$$
P = \dfrac{W}{t}  = \dfrac{W-0}{t-0} = \dfrac{\varDelta W}{\varDelta t}
$$
と前後の変化($\varDelta$)を意識した量で書くことが望ましいが、教科書では簡略化している。
