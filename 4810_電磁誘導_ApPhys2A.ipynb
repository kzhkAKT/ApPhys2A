#### ファラデーの電磁誘導の法則

以下、図4.58 については、
1. 一定の強さ磁束密度 $B$(向きは上向のまま)
2. 導線棒含むコイル全体に一様に広がっている

とします. つまり
- 図中の $B$ は代表的な磁束密度を表しています。

##### p98, 図4.58 (a) 外部回路/非接触

外部回路に繋がれていない(接触のない)、一定磁束密度 $B$ を移動する、導体(金属棒)を考える。[図(a)]
1. 図の右側へ導体棒を移動
2. ローレンツ力 $F$ が働き、導体内で電子が移動
3. 電子の移動により、電界 $E$ が生じる $(+,-)$
4. クーロン力により、最初の移動とは逆向きの力が生じる
5. 最終的には自由電子の動きが止まる  
   (移動できる自由電子がなくなる)

導体棒を軸とした力のつり合い:
\begin{align}
(-e)\,v\,B + e\,E = 0
\end{align}

これを、電界 $E$ について整理すると
\begin{align}
E = v\,B
\end{align}

右向きへ動く導体には、電界 $E$ が生じたと言える。

---

端的に言えば、導体棒が電池の役割を担うことになる

##### p98, 図4.58 (b) 外部回路/接触

1. 導体棒内の移動した電子が連続的に外部回路に供給される
2. 導体棒内に連続的に、電子の移動とは逆向きに、連続的に電流が流れる

電界 $E$ と 電位(電圧) $V$ の関係は
- $E$ は局所的な傾き
- $V$ は基準との高低差

を表すので、

- 導体棒の長さを $L$ とすると

> \begin{align*} E = \dfrac{V}{L}\end{align*}

が成り立つ。

##### 誘導起電力

\begin{align}
V = E\,L = (v\,B)\,L = v\,B\,L
\end{align}

導体棒を引く速度 $v$ は、速度の定義より: $v = \dfrac{\Delta x}{\Delta t}$



\begin{align}
V &= v\,B\,L
= B \dfrac{(L \Delta x)}{\Delta t}
= B \dfrac{\Delta(L\,x)}{\Delta t}\\
&= B \dfrac{\Delta S}{\Delta t}
= \dfrac{\Delta(B\,S)}{\Delta t}\\[2mm]
V &= \dfrac{\Delta \varPhi}{\Delta t}
\end{align}

---

ここで $\Delta\varPhi = \Delta(BS)$ として、導体棒の移動に伴う、  
コイルが貫く磁束 $BS$ の増加を表現した

このことから
> コイルに生じる誘導起電力は、コイルを貫く磁束の時間変化で与えられる

ことがわかり、これを **「ファラデーの電磁誘導の法則」** という

Lenz の法則
> **誘導電流はコイルを貫く磁束の変化を妨げる向きに流れる**

\begin{align*}
V &= - \dfrac{\Delta \varPhi}{\Delta t},\\

\end{align*}

この実験では、 $S$ は変わらず、   
$B$ が時間的に変化していることに注意

\begin{align*}
&\begin{cases}
I_{01} = 0 \to I_{1} \neq 0\\
I_1 \to B_1 \sim B_2 \to \varPhi_2= B_2\,S \to V_2 = -\dfrac{\Delta\varPhi_2}{\Delta t}\\[2mm]
\varPhi_2 \propto I_1  \to \Delta \varPhi_2  \propto \Delta I_1
\end{cases}\\[3mm]
V_2 &= -\dfrac{\Delta\varPhi_2}{\Delta t} \propto - \dfrac{\Delta I_1}{\Delta t} \\[5mm]
V_2 &= -\dfrac{\Delta\varPhi_2}{\Delta t} = - M \dfrac{\Delta I_1}{\Delta t}
\end{align*}




1. コイル1 に電流が流れ磁束密度 $B_1$ が変化
2. $B_1$ の変化を、距離の影響を受け、コイル2 が $B_2$ として受け取る
3. コイル2 の磁束 $\varPhi_2=B_2\,S$ が変化する
4. コイル2に、コイル1からの影響を受けた、相互誘導が生じる




- 比例定数 $M$ を 「コイル1,2の」 **相互インダクタンス** (mutual inductance)
- $M$ の単位は, 上の式の踏まえて, ${\rm [\frac{V}{A/s}] = [V.s/A]}$
  - この単位を 発見者の [Joseph Henry](https://en.wikipedia.org/wiki/Joseph_Henry) の名前にちなんで H,  **ヘンリ** という

\begin{align}
&\begin{cases}
I_0 =0 \to I \neq 0\\
B = \mu_0 \, H = \mu_0 \, \dfrac{I}{2\,r} \propto I
\end{cases}\\[2mm]
\varPhi
&= B\,S \propto B \propto I\\

\end{align}

$B\, S$ の $S$ はリングの面積

\begin{align}
V

&= - L \frac{\Delta I}{\Delta t}
\end{align}

1. 電流 $I$ が変化しているため、磁束密度 $B$ も変化
1. これによりコイル自身を貫く磁束 $B\,S$ も変化
1. よって **自己誘導** が起こる

---


$L$ はコイルに固有の量で、時間変化しない。


- 比例定数 $L$ を「注目する」コイルのみの **自己インダクタンス** (self inductance)
- $L$ の単位は、(式の形が同じなので) $M$ と同じ、henry

### (4) コイルのエネルギー

[分割表示によるコイル(に蓄えられた)エネルギー計算](https://kzhkakt.github.io/gadgets/inductor-energy-simulator/)

これまで学んだ関係式 [p61(4.7), p75 (4.24), p101 (4.54)]を、  
電流が定常流(一定な電流)になる途中に合わせる
\begin{align*}
\require{cancel}
\begin{cases}
V = \dfrac{\bcancel{U}\,{W}}{q} &\to \Delta W = \Delta(q\,V) = \Delta q\bcancel{V}\,{v} = \Delta q \cdot v\\[1mm]
I = \dfrac{\Delta Q}{\Delta t}  &\to i = \dfrac{\Delta q}{\Delta t}\\
V = - L \dfrac{\Delta I}{\Delta t}
&\to
v = - \left( - L \dfrac{\Delta i}{\Delta t}\right)
= + L \dfrac{\Delta i}{\Delta t}
\end{cases}
\end{align*}

---

３番目の式は、**誘導起電力** に「逆らって」コイルにする仕事なので、符号が変わっています。

途中の電流 $i$, そのときの電圧 $v$ 用いて、そのとき、その瞬間にコイルに逆らってする仕事 $\Delta W$ は

\begin{align}
\require{cancel}
\Delta W &= \Delta q\,v \\
&= (i\, \Delta t)\,v
= i\, \bcancel{\Delta t}\,\,L \dfrac{\Delta i}{\bcancel{\Delta t}} \\
&= L\,i\,\Delta i
\end{align}

全ての瞬間は $i=0$ から $i=I$ までを足し合わせたもの[上図の左(b)の短冊群]なので

\begin{align*}
W = \sum_{i=0}^{i=I} \Delta W = \sum_{i=0}^{i=I} L\,i\,\Delta i
\end{align*}

$\Delta i \to 0$ とすることで、グラフの隙間を埋めると、  
定積分の定義となり[三角形の面積に相当]

\begin{align*}
W &= \int_{i=0}^{i=I} d\,W = \int_{i=0}^{i=I} L\,i\,d\,i \\[3mm]
&= \dfrac{1}{2}\, L\, I^2
\end{align*}

これが、この回路を閉じた際に、コイルに蓄えられるエネルギーとなる。
