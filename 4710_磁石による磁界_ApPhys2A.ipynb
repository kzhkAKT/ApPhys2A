#### 電荷と磁荷と同じところ

電荷(+,-)と同じように磁荷(N,S)の間に引力や斥力が働く
- (充分小さく大きさの無視できる)磁荷大きさ $|m_1|, |m_2|$ に比例
- 距離の自乗に反比例 $\propto \frac{1}{r^2}$

距離 $r$ 離れた、(点)磁荷 $m_1, m_2$ に働く磁気的なクーロン力 $F_\mathrm{m}$

\begin{align*}
|F_m|=k_{\mathrm{m}} \frac{\left|m_1\right|\left|m_2\right|}{r^2}
\end{align*}

$k_{\mathrm{m}} = 6.33\times 10^4~\rm{N\,m^2/Wb^2}$(真空中).

---
距離 $r$ 離れた、(点)電荷 $q_1, q_2$ に働くクーロン力

\begin{align*}
|F_\mathrm{c}|=k \frac{\left|q_1\right|\left|q_2\right|}{r^2}
\end{align*}

$k = 9.00\times 10^9~\rm{N\,m^2/C^2}$(真空中).

外部の一様な電界(電場)ベクトル $\vec{E}$ に、点電荷 $q$ が存在する時のクーロン力 $\vec{F}_{\mathrm{c}}$ は

\begin{align*}
\vec{F} = q\,\vec{E}
\end{align*}

これは読み替えると $\vec{E} = \frac{\vec{F}}{q}$ となり
> $\vec{E}$ は単位電荷 $1~\rm{C}$ に働く力  

と読める。

---
外部の一様な磁界(磁場)ベクトル $\vec{H}$ に、(点)磁荷 $m$ が存在する時の磁気的な クーロン力 $\vec{F}_{\mathrm{m}}$ は、電界のアナロジーを利用して

\begin{align*}
\vec{F}_{\mathrm{m}} = m\,\vec{H}
\end{align*}

これは読み替えて $\vec{H} = \frac{\vec{F}_\mathrm{m}}{m}$ となり
> $\vec{H}$ は単位電荷 $1~\rm{Wb}$ に働く力  

と読める。

磁界ベクトル $\vec{H}$
-  $\left|\vec{H}\right| = H$ を磁界ベクトルの大きさ
- $\frac{\vec{H}}{\left|\vec{H}\right|}= \frac{\vec{H}}{H} = \vec{e}_{H}$ 磁界ベクトルの向き
- 単位は N/Wb

p87 青い囲み: 磁力線の性質
> 1. 磁力線の接線の向きは、その接点での磁界ベクトルの向きを向く。
> 1.   磁力線は、磁界が強い場所では密であり、磁界が弱い場所では疎である。
> 1.   磁力線は、磁荷のない場所で発生したり消滅したりしない。また、途中で折れ曲がったり、枝分かれしたり、交差したりもしない。

---

p57 青い囲み: 電気力線の性質
> 1. 電気力線の接線の向きは、その接点での電界ベクトルの向きを向く。
> 1. 電気力線は、電界が強い場所では密であり、電界が弱い場所では疎である。
> 1. 電気気力線は、正電荷から出て負電荷に入るか、正電荷から出て無限遠に行くか、無限遠から来て負電荷に入るかのいずれかである。
> 1. 電気力線は、電荷のない場所で発生したり消滅したりしない。また。途中で折れ曲がったり、枝分かれしたり、交差したりもしない。
