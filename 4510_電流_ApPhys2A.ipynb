断面積 = Area = **S**ection

---
from [https://courses.lumenlearning.com/](https://courses.lumenlearning.com/suny-physics/chapter/20-1-current/0)


\begin{align*}
I
&= \dfrac{Q}{t}\\[2mm]
&= \dfrac{Q - 0 }{t - 0} = \dfrac{\Delta Q}{\Delta t}
\end{align*}

時間 $t$ の間に断面積 $S$ を通過する 総電気量 $Q$ は
- 電子（またはキャリア）の数密度 $[{\rm 1/m^3}]$ : $n$
  - 電気素量（電子1個の持つ電荷の大きさ） : $e$
- 定常流として一定とみなした、電荷の(平均)速度 $[{\rm m/s}]$: $v$
- 導線の断面積 $[{\rm m^2}]$ : $S$
   - $t$ 秒に進める体積 $V= S\,\ell=  S\,(v\,t)$
\begin{align*}
Q &= e\, (n \cdot V) \\
&= e\,n\,(S\,\ell) =  e\,n\,S\,(v\,t)
\end{align*}

よって電流の定義より
\begin{align*}
I
&= \dfrac{Q}{t} = \dfrac{e\,n\,S\,v\,t}{t} = e\,n\,S\,v
\end{align*}

---

電子の移動方向と電流の向きは逆になることに注意しよう。（図の正電荷 $q$ の動きが電流の向きとなる）
