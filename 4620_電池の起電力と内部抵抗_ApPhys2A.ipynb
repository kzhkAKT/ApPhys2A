## 4.6.2 電池の起電力と内部抵抗

電池も電荷を通す素材で出来ている為、電流が流れる時には、抵抗を生じます。  
このように電池内部に存在する抵抗を、**内部抵抗** といいます。


```python
#@title　起電力 $E$ と 内部抵抗 $r$
with schemdraw.Drawing(mathfont='cm') as d:
#  d += elm.Line().down()
  elm.BatteryCell().down().label("$E$")
  elm.Resistor().down().label('$r$')
  elm.Line().right()
  elm.Line().up()
  elm.Resistor().up().label('$R$').label(['-','$V$','+'], loc='bot')
  L1=elm.Line().left()
  elm.CurrentLabel(top=False, ofst=.3, reverse=True).at(L1).label('$I$')
```

電池の正味の電圧 **起電力** $E$, 端子の抵抗(R)に加わる端子電圧を $V$ とする。

電圧の上昇と下降の関係から
$$
E = V + r\,I
$$
これを、端子に加わる電圧(端子電圧) $V$で
表現すると
$$
V = E - r\,I
$$


端子で成り立つオームの法則 $V=R\,I$ を代入すると
$$
E = R\,I + r\,I,
$$
これを利用することで、この回路に流れる電流 $I$ が次の様に表現できる
$$
I = \dfrac{E}{R+r}
$$
