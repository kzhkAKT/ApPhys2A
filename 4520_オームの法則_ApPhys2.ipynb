### 抵抗率 $\rho = \rho(t)$ の温度計数 $\alpha$

$\require{cancel}$
- $\rho_0$  at $0 \, {}^\circ{\rm C}$

at $t \, {}^\circ{\rm C}$
$$
\varDelta \rho (t) := \rho(t) - \rho_0
\propto
\begin{cases}
\rho_0\\[2mm]
\varDelta t = (t - 0) \, {}^\circ{\rm C}
\end{cases}
$$

$$
\begin{align*}
\varDelta \rho(t) &= \alpha\, \rho_0 \, \varDelta t\\[2mm]
\rho(t) - \rho_0 &= \alpha\, \rho_0 \, \varDelta t\\[5mm]
\rho(t) &= \rho_0 \left( 1 + \alpha\,\varDelta t\right)
\end{align*}
$$

抵抗率の温度計数 $\alpha$ の単位
$$
\left[ \alpha \right]
= \left[\dfrac{\varDelta \rho}{\rho_0\, \varDelta t}\right]
= \left[\dfrac{\bcancel{\varDelta \rho}}{\bcancel{\rho_0}\,\varDelta t}\right]
= \left[\dfrac{1}{t}\right]  = 1/{}^\circ{\rm C} = 1/{}^\circ{\rm K}
$$
