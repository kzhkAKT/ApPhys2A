### (1) 交流の発生

動的確認
- https://javalab.org/en/ac_generator_en/




時刻 $t$ での磁束 $\varPhi(t)$, 磁束密度$\times$面積, は
\begin{align*}
&\begin{cases}
S(t) &= S~\cos~\omega\,t\\
\varPhi(t) &= BS(t)
\end{cases}\\[3mm]
\varPhi(t) &=  B\,S\cos~\omega\,t
\end{align*}

$\Delta t$ が十分小さいことを考慮し、整理すると

\begin{align*}
\Delta \varPhi(t)
&= \varPhi(t + \Delta t) - \varPhi(t)\\
&= BS \left( \cos(\omega(t + \Delta t)) - \cos(\omega\,t)\right)\\
&= BS~\Delta t
~ \dfrac{ \cos(\omega(t + \Delta t)) - \cos(\omega\,t)}{\Delta t}\\
&= BS~\omega\Delta t
~ \dfrac{ \cos(\omega\,t + \omega\Delta t) - \cos(\omega\,t)}{\omega\Delta t}\\
&= BS\omega~\Delta t \left(  - \sin (\omega\,t)\right)\\[2mm]
%
\Delta \varPhi(t)
&= - \Delta t ~BS\omega\sin \omega\,t
\end{align*}

磁束の変化 $\varPhi$ が、状況を元に戻そうと誘導機電力が働くので、Lenz の法則より
\begin{align*}
V = - \dfrac{\Delta \varPhi}{\Delta t} = BS \omega \sin\omega\,t
\end{align*}
正弦(sine)の最大値(+1)と最小値(-1)を考え, $V_0=BS\omega$ とおけば
\begin{align*}
V = (BS \omega) \sin\omega\,t =V_0 \sin(\omega\,t )
\end{align*}

このことから
> 磁束中を回転する面積によって、時間的に変動する **交流電圧**　が発生した

と表現することができる。




*   $\omega$ はリングコイルを回転させる角速度であったが、  
交流では単振動と同じ電圧を発生しているため、$\omega$ を **角周波数** と呼ぶ.
\begin{align*}
\omega = \dfrac{2\pi}{T} = 2\pi~\dfrac{1}{T} = 2\pi~f
\end{align*}
*   $f$ はリングコイルの回転数を表すが、  
出力電圧が単振動的に時間変動するため、$f$ を **周波数**と呼ぶ.

\begin{align*}
f = \dfrac{1}{T} = \dfrac{\omega}{2\pi}
\end{align*}

### (2) 抵抗に流れる電流

交流を考え得る際にも、抵抗 $R$ は固有の値で時間変動の影響を受けないので、  
オームの法則 $V=I\,R$ がそのまま利用でき、電流 $I$ は次のようにもとまる

\begin{align*}
I
&= \dfrac{V}{R} = \dfrac{V_0~\sin\omega\,t}{R} \\
&= \dfrac{V_0}{R}~\sin\omega\,t = I_0~\sin(\omega\,t)
\end{align*}

- $I_0=\frac{V_0}{R}$ は交流**電流**の最大値
- $\sin$ の括弧の中は、**位相** と呼ばれる
- 交流電圧 $V$ と 交流電流 $I$ は **同位相** である

#### 抵抗で消費される電力 $P$

電力 $P$ (Power)は直流で考察した関係をそのまま交流にも適用する

\begin{align*}
P
&= V\,I = V_0\sin(\omega\,t )\, I_0\sin(\omega\,t)\\[2mm]
&= V_0\,I_0 \sin^2 \omega\,t
\end{align*}

- $P$ も時間的に変動する
- $P \geqq 0$



```python
#@title 交流電圧$V$(青) と電流$I$(赤) と電力$P$(オレンジ) の図示
#@markdown $y$ 軸の単位が異なることに注意
#@markdown 紫色の点はオレンジ色の曲線の最大値と最小値の平均
%%html
<iframe src="https://www.desmos.com/calculator/i8r10qacil?embed" width="500" height="500" style="border: 1px solid #ccc" frameborder=0></iframe>
```


$P-t$ グラフの形状を踏まえると、一周期 $T$ あたりの 電力 ${P}$ 平均値 $\bar{P}$ は  
(オレンジの塗りつぶし領域に相当するので)
\begin{align}
\bar{P}
&=
\dfrac{\frac{1}{2}~V_0\,I_0~T}{T}
= \dfrac{V_0\,I_0}{2}\\[2mm]
&=
\dfrac{RI_0^2}{2} = \dfrac{V_0^2}{2R}
\end{align}

- 直流の場合とは係数 $\frac{1}{2}$ だけ異なる
- 直流の場合と同様の表記にするため **実効値** (effective value) を導入

\begin{align}
&\begin{cases}
V_e&=\dfrac{V_0}{\sqrt{2}}\\[2mm]
I_e&=\dfrac{I_0}{\sqrt{2}}
\end{cases}\\[5mm]
%
\bar{P}
&=
 \dfrac{V_0\,I_0}{2}
= \dfrac{V_0}{\sqrt{2}}\, \dfrac{I_0}{\sqrt{2}}\\[2mm]
&=
V_e I_e= RI_e^2 = \dfrac{V_e^2}{R}
\end{align}
となり、**実効値** を利用すると直流による表現と同じ形式が得られる。



* 家庭用電源の $V_e=$100V は **実効値**
* 刻々と変化する時間 t を変数にもつ、 $V=V(t), I=I(t)$ を **瞬時値**

* $V_0$ や $I_0$ は、それぞれ *交流電圧最大値* と*交流電流最大値*



1.  発電所で発電する一定電力 $P$
  - その出力(交流)電圧はで $V$ とする
2.  送電線で失われるジュール熱 $p$
  - 送電線自体には、一定の抵抗 $r$ のみとする
3.  流れる電流は、発電所と送電線は繋がっているので、 $I$
4. 発電所(出力)電力 $P$ と送電線(損失)ジュール熱 $p$
  - 送電時の損失比 $p/P$ を調べる



発電所では $V,I$ のみを送るとし抵抗は考えないとすると,
発電所(出力)で発生する仕事率(ジュール熱) $P$ は
$$
P = V\,I.
$$
一方、送電線の抵抗があるため、その抵抗を $r$ として,
送電線で発生するジュール熱 $p$ は
\begin{align*}
&\begin{cases}
p &= V\,I = (r\,I)\,I = r\,I^2\\
&V = r\,I
\end{cases}\\[2mm]
\end{align*}

出力を基準($P=IV$)に、これらの比を取ると
\begin{align*}
\dfrac{p}{P}
&= \dfrac{r\,I^2}{V\,I}
= \dfrac{r}{V}\,I
= \dfrac{r}{V}\,\dfrac{P}{V}
= {r}{P} \dfrac{1}{V^2}
\end{align*}

- $P$ と $r$ が一定ならば, $p/P$ は $V^2$ に反比例
- つまり $V$ が大きければ、損失比 $p/P$ を小さくできる

### (5) コイルに流れる交流 : (V-)L回路

(交流) 電圧とコイルを持つ、(V-)L回路図を見ながら、  
KVL(キルヒホッフの電圧則)を適用すると
\begin{align}
V + V_{L} = 0
\end{align}

Lenz の法則より
\begin{align}
V_{L} = \left( - L \dfrac{\Delta I}{\Delta t} \right)
\end{align}

整理すると
\begin{align}
V + \left( - L \dfrac{\Delta I}{\Delta t} \right) = 0
\end{align}
(図の)交流電圧$V = V_0 \sin \omega t$ を代入すると
\begin{align}
V_0 \sin \omega t + \left( - L \dfrac{\Delta I}{\Delta t} \right) = 0
\end{align}
さらに
\begin{align}
\dfrac{\Delta I}{\Delta t}
=
\dfrac{V_0}{L}\, \sin \omega t
\end{align}

$\Delta t \to 0$ とすると、微分として表現されるので、   
積分領域に注意して $d\,t$ をかけて積分すると
\begin{align*}
\require{cancel}
\dfrac{d\,I}{d\,t} &= \dfrac{V_0}{L}\, \sin \omega\,t\\[1mm]
\int_{I(0)}^{I(t)} \dfrac{d\,I}{\bcancel{d\,t}}\,\bcancel{d\,t}
 &= \int_{t=0}^{t > 0} \dfrac{V_0}{L}\, \sin \omega\,t\,d\,t\\[1mm]
%
I(t) - I(0)
&=
\dfrac{V_0}{L\omega} \left[ -\cos \omega\,t \right]{}_{t=0}^{t > 0}\\[1mm]
I(t) - I(0)
&=
\dfrac{V_0}{L\omega} \left( -\cos \omega\,t + 1 \right)
\end{align*}


時間に依存しない項を、次のように選ぶと、初期状態の影響を忘れることができる
\begin{align*}
-I(0) = + \dfrac{V_0}{L\omega}.
\end{align*}

これより、
\begin{align*}
I(t)
&= - \dfrac{V_0}{L\omega} \cos \omega\,t
= - I_0 \cos \omega\,t\\[2mm]
&= I_0  \sin \left( \omega\,t - \dfrac{\pi}{2} \right)
\end{align*}
ここで、$I_0 = \dfrac{V_0}{L\omega}$



交流電流 $V$ から得られる、VL回路上の交流電流 $I$ をまとめると
\begin{align}
\begin{cases}
V = V(t) &= V_0 \sin \omega t\\[1mm]
I = I(t) &= I_0 \sin \left(\omega t - \dfrac{\pi}{2} \right)
\end{cases}
\end{align}

これより、VL回路では
- $I(t)$ は $V(t)$ より、$-\pi/2$ ずれる、つまり
> **位相として $\pi/2$  遅れる** ことになる
- $I_0 = \dfrac{V_0}{L\omega}$ は、$V_0$ を元に $I_0$ が生じた為
> VL回路において $V_0$ から $I_0$ を生じるに  
必要な **$L\omega$ をコイルによる誘導リアクタンス** と呼ぶ
- 誘導リアクタンスは、単位は次のように考えれば $\Omega$ となる
\begin{align*}
V_0 = R_{L} I_0 = \left( L \omega \right) I_0
\end{align*}

### (6) コンデンサーに流れる交流: (V-)C 回路

コンデンサ(capasitor)は電荷 $Q$ を蓄える素子で、その容量を $C$ とすると、  
コンデンサ 両端の電圧を $V$ とすると
\begin{align*}
Q = C\,V
\end{align*}

```python
# @title 直流回路におけるコンデンサの動作
%%html
<iframe src="https://phet.colorado.edu/sims/html/capacitor-lab-basics/latest/capacitor-lab-basics_en.html"
        width="800"
        height="600"
        allowfullscreen>
</iframe>
```

この $C$ に、時間的に変動する交流電圧 $V=V(t)=V_0\,\sin\omega\,t$ を加えると   
蓄えられる電荷の容量 $Q$ は
\begin{align*}
Q(t) = Q = C\,V = C\,V_0\,\sin\omega t
\end{align*}

電流 $I$ は単位時間あたりの電荷の変化 $\Delta Q/ \Delta t \sim d\,Q/d\,t$ なので
\begin{align*}
I(t)
&= \dfrac{d\,Q(t)}{d\,t}= C\, \dfrac{d\,V}{d\,t}\\[2mm]
&= C\,V_0\,\omega\,\cos\omega t = I_0 \, \cos\omega t
\end{align*}

ここで $I_0 = C\, V_0 \,\omega$ とした。

交流電流 $V$ から得られる、VC回路上の交流電流 $I$ をまとめると
\begin{align}
\begin{cases}
V = V(t) &= V_0 \sin \omega t\\[1mm]
I = I(t) = I_0 \cos \omega t &=  I_0 \sin \left( \omega t + \dfrac{\pi}{2} \right)
\end{cases}
\end{align}

これより、VC回路では
- $I(t)$ は $V(t)$ より、$+\pi/2$ ずれる、つまり
> **位相として $\pi/2$  進む** ことになる
- $I_0 = C\, V_0 \,\omega =\dfrac{V_0}{\frac{1}{C\omega}}$ は、$V_0$ を元に $I_0$ が生じた為
> VC回路において $V_0$ から $I_0$ を生じるに  
必要な **$\dfrac{1}{C\omega}$ をコンデンサによる容量リアクタンス** と呼ぶ

- 容量リアクタンスは、単位は次のように考えれば $\Omega$ となる
\begin{align*}
V_0 = R_{C} I_0 = \left( \dfrac{1}{C\omega} \right) I_0
\end{align*}

### (7) (V-)RLC直列回路

1. この直列回路には共通の流れとして、交流電流 $I=I(t)$ が流れる。  
それを次のように仮定する
\begin{align*}
I = I(t) = I_0\,\sin \omega t
\end{align*}
2. $I(t)$ を利用しながら、位相の変化も踏まえて、 $R, L ,C$ での交流電圧を表記する
- at $R$,  
\begin{align*}
V_R = R\,I = R\, I_0\,\sin \omega t
\end{align*}
- at $L$,
\begin{align*}
V_L = R_L\,I_L
= (L\omega) \, I_0\,\sin \left( \omega t + \dfrac{\pi}{2} \right)
\end{align*}
- at $C$,
\begin{align*}
V_C = R_C\,I_C
= \left(\dfrac{1}{C\omega}\right)\,I_0\,\sin \left( \omega t - \dfrac{\pi}{2} \right)
\end{align*}

直列回路なので総合素子間の(交流)電圧の合計は、電源電圧 $V$　に等しいので
\begin{align*}
V
&= V_R + V_L + V_C\\
&=
R\, I_0\,\sin \omega t
+ (L\omega) \, I_0\,\sin \left( \omega t + \dfrac{\pi}{2} \right)
+ \left(\dfrac{1}{C\omega}\right)\,I_0\,\sin \left( \omega t - \dfrac{\pi}{2} \right)\\
&=
R\, I_0\,\sin \omega t
+
L\,\omega\,I_0\,\cos \omega t
-
\dfrac{I_0}{C\,\omega}\,\cos \omega t\\
&=
I_0
\left\{
  R\,\sin\omega t
  +
  \left( L\omega - \dfrac{1}{C\omega}\right) \cos \omega t
\right\}\\
&=
I_0\,Z\,\sin(\omega t + \phi)
\end{align*}

ここで
\begin{align*}
\begin{cases}
Z &= \sqrt{R^2 + \left( L\omega - \dfrac{1}{C\omega}\right)^2}\\
\tan\phi &= \dfrac{L\omega - \dfrac{1}{C\omega}}{R}
\end{cases}
\end{align*}

- 形式的には $V_0 = Z\,I_0$
- **$Z$ をインピーダンス**
- **$\phi$ を位相角**

image.png

1. $ L\omega - \dfrac{1}{C\omega}=0$ で、$Z$ は最小値
  - このとき $\omega = \dfrac{1}{\sqrt{LC}}$
2. さらに、$V$ が供給される一定電圧と考えれば、 $I$ は最大を取りうる
  - この回路の現象は、力学の共振と似ているため **共振回路** と呼ばれる
    - $\omega = \dfrac{1}{\sqrt{LC}}$ を共振角周波数
    - $f_0 = \dfrac{\omega}{2\pi} =  \dfrac{1}{2\pi\sqrt{LC}} $ を **共振周波数**
