上の実験から「電流 $I$ と 磁界 $H$ と力$F$」のそれぞれに方向があるので、  
次のように **ベクトル** 量して考えて行きます。
> - 電流(ベクトル) $\vec{I}$ と 磁界(ベクトル) $\vec{H}$ と力(ベクトル)$\vec{F}$

とします。

- 力$F$によって、導線分 $\ell$ も 動く

も留めておいて下さい。


もう一歩踏み込むと

> 磁荷 $m$ を持たない $\vec{I}$ と $\vec{H}$ の相互作用が $\vec{F}$ を発生させている.

ので

> $\vec{I}$ のもつ、電荷 $q$ の流れ(速度)
 $\vec{v}$ と 相互作用することで、
 導線分 $\ell$ に
 $\vec{F}$ を発生させる磁力線

が考えられます。

### 磁界 $\vec{H}$ と 磁束密度 $\vec{B}$ との関係
\begin{align*}
\begin{cases}
(m, \vec{H}) &\to \vec{F}_m &\quad \text{磁荷によるクーロン力}\\
\left(\vec{I},  \vec{B} \right) \sim \left((q,\vec{v},\ell),  \vec{B}\right) &\to \vec{F}_q &\quad \text{電流}~\vec{I}~\text{によるクーロン力}
\end{cases}
\end{align*}
$\vec{H}$, $\vec{B}$ も磁力線を物理的に扱うために表現したものなので
> $\vec{B} \propto \vec{H}$

と並行なはずです。
そこで等式にするためには

\begin{align*}
\vec{B} = \mu_0 \, \vec{H}
\end{align*}

$\mu_0$ は、真空中の透磁率(permeability) と呼ばれる量で
> $\mu_0 = 4\pi \times 10^{\,-7}~{\rm N/A^2} = 1.26 \times 10^{\,-6}~{\rm N/A^2} $

で、これを踏まえると $\left| \vec{B} \right| = B$ の単位は
>  ${\rm (N/A^2)(N/Wb) = (N/A^2)(A/m) = N/(Am)= (N\,m/A)/m^2= (J/A)/m^2 = Wb/m^2}$

と変換され、「単位面積あたりの磁力線の強さとして」
> ${\rm T}$ (テスラ)

で表現される。

---

真空ではない、一般の場合には磁力線が存在する物質の影響を、
物質の透磁率 $\mu$ で表現し
\begin{align*}
\vec{B} = \mu \, \vec{H}
\end{align*}

となります。


### $\vec{I}, \vec{B} , \vec{F}$ の方向確認

- $\vec{F}\, \bot\, \vec{B}$
- $\vec{F}\, \bot\, \vec{I}$

です。

![フレミングの左手の法則](https://upload.wikimedia.org/wikipedia/commons/3/32/Fleming%27s_Left_Hand_Rule.png)



### 導線分 $\ell$ に働く力 $\vec{F}$

外積/ベクトル積記号 $\times$ を用いて
\begin{align}
\vec{F} = \ell \vec{I} \times \vec{B}
\end{align}
その大きさは
\begin{align}
\left| \vec{F} \right| &= F
= \ell \left| \vec{I} \right| \left| \vec{B} \right| \sin \alpha\\[2mm]
%
F &= \ell I\, B \, \sin\alpha
\end{align}

ここで、 $\alpha$ は、電流(ベクトル) $\vec{I}$ と 磁束密度(ベクトル) $\vec{B}$ がなす角度です。

---

- 左手の人差し指と中指がなす角度が $\alpha$
- 力の方向は、一般的な(右)ネジが進む方向

#### 長方形コイルが受ける力のモーメント

* このようなコイルは中心を固定されている為、回転します。
* 回転を産む力が、中心から離れているため、トルク=力のモーメントが生じます
* 力のモーメントは
\begin{align*}
\vec{\tau} = \vec{r} \times \vec{F}
\end{align*}
で表現されます。
  * 上の図では、<font color="green">緑色で囲まれた平行四辺形の面積</font>に相当します。  
  (平行四辺形は右上と左下二箇所にあります。)
  * 上下の緑色の平行四辺形の $\left| \vec{\tau} \right|$ は、  
  それぞれコイルを固定された軸を中心に、**反時計回りに**回すことになります。

> 回転軸がPQ,RSに平行に固定されているため、  
> PS, QR で生じる力は固定されたコイルの回転に寄与しません

\begin{align*}
\left| \vec{\tau} \right|
= N
&= (1+1) \, \left| \vec{r} \times \vec{F}\right|
= 2 \, \dfrac{b}{2} \, F \sin(\frac{\pi}{2}-\theta)\\
&= b\,F\,\cos \theta
\end{align*}

$\vec{F}$ の大きさ $F$ については
\begin{align}
\vec{F}
&= L\, \vec{I} \times \vec{B}\\
|\vec{F}| = F &= a\,I\,B\,\sin(\frac{\pi}{2}) = aIB
\end{align}

これらをまとめると
\begin{align*}
\left| \vec{\tau} \right|
= N
&=b\,(aIB)\,\cos \theta \\[2mm]
&= abIB \cos\theta
\end{align*}

### (4) ローレンツ力

磁束密度 $\vec{B}$ の中を流れる電流 $\vec{I}$ に働く力 $\vec{F}$ の大きさは

\begin{align}
\left| \vec{F} \right| = F
&= \left|L\, \vec{I} \times \vec{B} \right|
= L\,I\,B\,\sin \theta\\
&= L\,\left(\dfrac{d\,Q}{d\,t}\right)\,B\,\sin \theta\\
&= L\,\left(\dfrac{d\,(n\,e\,S\,x)}{d\,t}\right)\,B\,\sin \theta\\
&= L\,n\,e\,S\,v\sin\theta
\end{align}

これを素電荷 $e$ あたりに換算すると
\begin{align}
f = \dfrac{F}{N_Q} = \dfrac{F}{n\,S\,L} = e\,v\,B\,\sin\theta
\end{align}


素電荷 $e$ に $\pm$ を加えることで、一般的な電荷 $q$ を表現できることを踏まえると

> 電荷 $q$ に働く **ローレンツ力** の大きさは $f$
\begin{align}
f = \left| q\right|\, v\, B\, \sin\theta
\end{align}

これを外積 $\times$ を利用して表現すると
\begin{align*}
\vec{f} = q\,\vec{v} \times \vec{B}
\end{align*}

これを利用すると、速度 $\vec{v}$ を持つ電荷 $q$ が、  
磁束密度 $B$ に入ると下の図のように曲がることが表現できる

![](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Lorentz_force.svg/589px-Lorentz_force.svg.png)

#### 教科書の式導出

![](https://upload.wikimedia.org/wikipedia/commons/thumb/5/5b/Cyclotron_orbit.svg/250px-Cyclotron_orbit.svg.png)

##### (電子に働く)ローレンツ力

$$
\begin{align*}
\vec{f} &= q \vec{v} \times \vec{B} = (-e) \vec{v} \times \vec{B}\\[3mm]
|\vec{f}|
&=  |(-e)|~|\vec{v}|~|\vec{B}|~\sin\left(\frac{\pi}{2}\right)\\
&= e v B
\end{align*}
$$

---

- 電子の電荷: $(-e)$
- 電子の速度: $\vec{v}$
  - 電子の速さ: $v$
- 電子に印加する磁束密度: $\vec{B}$
  - その磁束密度大きさ: $B$


##### 等速円運動を発生させる向心力の見積($r_c=r$)



$$
f = m a = m (v \omega) = m v \left(\dfrac{v}{r_c}\right) = m \dfrac{v^2}{r_c}
$$

---

- 電子の質量: $m$
- 等速円運動する軌道の半径: $r_c$
- 等速円運動する物体の速さ: $v= r \omega$
  - $\omega$ は角速度
- 等速円運動する物体の加速度の大きさ: $a= v \omega$


なお、力ベクトルと速度ベクトルの内積は、物理量としては **仕事率** [W] に相当する。

$$
\vec{f}\cdot \vec{v} = \left( (-e) \vec{v} \times \vec{B} \right) \cdot \vec{v}= 0
$$

この仕事率が、常にゼロなので、エネルギーは増減しません。


##### ローレンツ力が向心力の源

$$
\require{cancel}
e v B = f = m \dfrac{v^2}{r_c}\\
e \bcancel{v} B = f = m \dfrac{v^\bcancel{2}}{r_c}\\
e B = m \frac{v}{r}\\
r_c = \dfrac{m v}{e B}
$$

このように導出することで
$$
r_c = r_c(m,v,e,B)
$$
と表現される

##### サイクロトン運動の周期 $T$

$$
v = \dfrac{\text{一周する距離}}{\text{一周する時間}}
= \dfrac{2 \pi r_c}{T}\\[3mm]
T = \dfrac{2\pi r_c}{v} =  2 \pi \dfrac{m}{eB}
$$

このように導出することで
$$
T = T (m,e,B)
$$
と表現される。
